```python
messy = pd.DataFrame({ ## messy.isnull().sum()
    "product" : ["pen","book","car","scooter"],
    "price" : [20,np.nan,30,np.nan],
    "rating" : [4.1,4.5,4.8,np.nan]
})
```

```python
messy
```

```python
messy.isnull().sum()
```

```python
#skewed --- median(eg - price)
# data is symmetric --- mean(eg - rating)
# categorical --- mode (eg - product name)
```

```python
cleaning = messy.copy()
```

```python
cleaning["price"] = cleaning["price"].fillna(cleaning["price"].median())
```

```python
cleaning["rating"] = cleaning["rating"].fillna(cleaning["rating"].mean())
```
